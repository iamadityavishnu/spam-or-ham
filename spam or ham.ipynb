```python
import pandas as pd
```

```python
data = pd.read_csv('owndata.csv', encoding="unicode_escape")
data.head()
```

```python
data = data.replace(to_replace = ["ham", "spam"], value = ["0", "1"])
data.head()
```

```python
data.describe()
```

```python
data.label.value_counts()[1]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
X = vectorizer.fit_transform([x.lower() for x in data["email"]])
```

```python
tfidf_tokens = vectorizer.get_feature_names()
```

```python
X = pd.DataFrame(X.toarray())
print(X)
```

```python
Y = X.set_axis(tfidf_tokens, axis=1, inplace=False)
print(Y)
```

```python
Y['label'] = data['label']
print(Y)
```

```python
p_dist = Y.groupby('label').sum()
```

```python
p_dist.head()
```

```python
p_dist += 1
p_dist.head()
```

```python
p_dist.loc['0'] = p_dist.loc['0'] / p_dist.max(axis=1)[0]
p_dist.loc['1'] = p_dist.loc['1'] / p_dist.max(axis=1)[1]
```
